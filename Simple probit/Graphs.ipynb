```python
from scipy.stats import t, norm, logistic, uniform
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.figsize'] = 12, 8
%precision 4
plt.style.use('_classic_test_patch')

dots = np.linspace(-20, 20, 1000)

yt = t.pdf(dots, df=20, loc=0, scale=10)
ynorm = norm.pdf(dots, loc=0, scale=10)
yu = uniform.pdf(dots, loc=-10, scale=20)
ylog = logistic.pdf(dots, loc=0, scale=20)
```

```python
plt.plot(dots, ynorm, label='$\mathcal{N(0, 10)}$')
plt.plot(dots, yu, label='$\mathrm{U(-10, 10)}$')
plt.plot(dots, ylog, label='$\mathrm{L(0, 20)}$')
plt.plot(dots, yt, label='$\mathrm{t(20, 0, 10)}$')
plt.title('Априорные распредления')
plt.xlabel('$\mathrm{x}$')
plt.legend();
plt.savefig('priors.pdf', dpi=200)
```
